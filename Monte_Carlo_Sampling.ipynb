```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def check_whether_point_on_circle(x: float ,y: float, r: float=1) -> bool:
  return x ** 2 + y ** 2 <= r ** 2
```

```python
def sampling(K: int, r: float=1) -> list[tuple]:
  assert K > 0

  pts = []

  for _ in range(K):
    x, y = np.random.rand(2) * 2 - 1
    pts.append((r * x, r * y))

  return pts
```

```python
def calc_pi(pts: list[tuple], r: float=1) -> float:
  ## Estimate the area of circle
  cnt = 0

  for p in pts:
    cnt += int(check_whether_point_on_circle(*p, r))

  ratio = cnt / len(pts)

  area = ((2 * r) ** 2) * ratio

  ## Compute the value of pi
  pi = area / (r ** 2)

  return pi
```

```python
def scatter_plot(pts: list[tuple], r: float=1) -> None:

  fig, ax = plt.subplots(1,1,figsize=(5,5))

  ## points
  X = [p[0] for p in pts]
  Y = [p[1] for p in pts]
  ax.scatter(X, Y, marker='.')

  ## circle
  step = 0.01
  X_circle = np.arange(-1, 1 + step, step) * r
  X_circle = np.clip(X_circle, -1, 1)
  U_circle = (r ** 2 - X_circle ** 2) ** 0.5
  L_circle = - (r ** 2 - X_circle ** 2) ** 0.5
  ax.plot(X_circle, U_circle, 'k-')
  ax.plot(X_circle, L_circle, 'k-')

  ## Estimation
  pi = calc_pi(pts, r)
  ax.set_title(f"PI = {pi:.6f}")

  ## axes design
  offset = 1.05
  ax.set_xlim(-r * offset, r * offset)
  ax.set_ylim(-r * offset, r * offset)

  plt.show()
```

```python
K = 100
r = 1.

## Sampling
pts = sampling(K, r)

## Visualization
scatter_plot(pts, r)
```
